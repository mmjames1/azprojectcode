```python
#import the modules
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# DL Tau

notes on first fit:

Ntot         : 9.4868000e+17 cm^-2

Gasrad       : 0.66700000 AU

Tgas         : 400.00000 K

Plaw_tem     : 0.25000000

Chi2 =  192.58396


```python
#import the spitzer IRS data
dataSH = fits.open('/Users/mackenzie/REDUCED/DLTAU_SH_final.fits')
dataLH = fits.open('/Users/mackenzie/REDUCED/DLTAU_LH_final.fits')
x = np.append(dataSH[1].data['WAVE'][0],dataLH[1].data['WAVE'][0])
y = np.append(dataSH[1].data['SPEC'][0],dataLH[1].data['SPEC'][0])
good = np.logical_not(np.isnan(y))
Lam_IRS = x[good]
Flux = y[good]

#Data from line fitting - continuum subtracted!
DLdata = pd.read_csv('Spectrum_DLTau.csv')
#lists from the columns
Lam = DLdata['Lam[micron]'].tolist()
Line_Only = DLdata['Only_Line[Jy]'].tolist()
Line_Cont = DLdata['Line+Cont.[Jy]'].tolist()
```

```python
#check plot
#plotting in the 3 panel format

fig, axs = plt.subplots(3, figsize=(10,8), sharex=False, sharey=False)

axs[0].plot(Lam,Line_Cont,drawstyle='steps-mid',color='red',label="Simulated Spectra")
axs[0].plot(Lam_IRS,Flux,drawstyle='steps-mid',linestyle='dashdot',color='black',label="Spitzer Spectra")
axs[0].set_xlim(11,18)
axs[0].set_ylim(0.8,1.1)
#fitting range
axs[0].axvspan(15.1, 15.3, color='silver')
axs[0].axvspan(15.5, 16.2, color='silver')
axs[0].axvspan(16.6, 16.9, color='silver')
axs[0].axvspan(17.0, 17.4, color='silver')

axs[1].plot(Lam,Line_Cont,drawstyle='steps-mid',color='red',label="Simulated Spectra")
axs[1].plot(Lam_IRS,Flux,drawstyle='steps-mid',linestyle='dashdot',color='black',label="Spitzer Spectra")
axs[1].set_xlim(19,25.7)
axs[1].set_ylim(1.0,1.3)
#fitting range
axs[1].axvspan(19.8, 20.0, color='silver')
axs[1].axvspan(20.4, 20.6, color='silver')
axs[1].axvspan(20.9, 21.1, color='silver')
axs[1].axvspan(21.6, 22.0, color='silver')
axs[1].axvspan(23.4, 23.7, color='silver')
axs[1].axvspan(23.8, 24.1, color='silver')
axs[1].axvspan(25.3, 25.4, color='silver')

axs[2].plot(Lam,Line_Cont,drawstyle='steps-mid',color='red',label="Simulated Spectra")
axs[2].plot(Lam_IRS,Flux,drawstyle='steps-mid',linestyle='dashdot',color='black',label="Spitzer Spectra")
axs[2].set_xlim(25.8,32)
axs[2].set_ylim(1.0,1.3)
#fitting range
axs[2].axvspan(25.9, 26.1, color='silver')
axs[2].axvspan(26.5, 26.6, color='silver')
axs[2].axvspan(28.3, 28.7, color='silver')
axs[2].axvspan(21.6, 22.0, color='silver')
axs[2].axvspan(29.1, 29.7, color='silver')
axs[2].axvspan(32.7, 33.3, color='silver')
axs[2].axvspan(33.4, 33.6, color='silver')


fig.suptitle("Comparison of Simulated and IRS Spectra for DL Tau",fontsize = 16)
fig.tight_layout()
fig.subplots_adjust(top=0.93)
legend = axs[0].legend(loc='upper right',fontsize='large')
```

```python
#comparing the line fitting

fig, axs = plt.subplots(3, figsize=(10,8), sharex=False, sharey=False)

axs[0].plot(Lam,Line_Only,drawstyle='steps-mid',color='red',label="DL Tau")
axs[0].set_xlim(11,18)
#fitting range
axs[0].axvspan(15.1, 15.3, color='silver')
axs[0].axvspan(15.5, 16.2, color='silver')
axs[0].axvspan(16.6, 16.9, color='silver')
axs[0].axvspan(17.0, 17.4, color='silver')

axs[1].plot(Lam,Line_Only,drawstyle='steps-mid',color='red',label="DL Tau")
axs[1].set_xlim(19,25.7)
#fitting range
axs[1].axvspan(19.8, 20.0, color='silver')
axs[1].axvspan(20.4, 20.6, color='silver')
axs[1].axvspan(20.9, 21.1, color='silver')
axs[1].axvspan(21.6, 22.0, color='silver')
axs[1].axvspan(23.4, 23.7, color='silver')
axs[1].axvspan(23.8, 24.1, color='silver')
axs[1].axvspan(25.3, 25.4, color='silver')

axs[2].plot(Lam,Line_Only,drawstyle='steps-mid',color='red',label="DL Tau")
axs[2].set_xlim(25.8,32)
axs[2].axvspan(25.9, 26.1, color='silver')
axs[2].axvspan(26.5, 26.6, color='silver')
axs[2].axvspan(28.3, 28.7, color='silver')
axs[2].axvspan(21.6, 22.0, color='silver')
axs[2].axvspan(29.1, 29.7, color='silver')
axs[2].axvspan(32.7, 33.3, color='silver')
axs[2].axvspan(33.4, 33.6, color='silver')


fig.suptitle("Continuum-Subtracted Spectra for DL Tau",fontsize=16)
fig.tight_layout()
fig.subplots_adjust(top=0.93)
legend = axs[0].legend(loc='upper right',fontsize='large')
```
