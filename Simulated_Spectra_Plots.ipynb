```python
#import the modules
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Linemod 21 

```python
Spitzer21 = pd.read_csv('Spitzer21_spectrum.csv')
JWST21 = pd.read_csv('JWST21_spectrum.csv')

#x data for all plots
S21wave = Spitzer21['Lam[micron]'].tolist()
J21wave = JWST21['Lam[micron]'].tolist()
```

```python
# spectra plot

S21 = Spitzer21['Line+Cont.[Jy]'].tolist()
J21 = JWST21['Line+Cont.[Jy]'].tolist()

#plotting in the 3 panel format
fig, axs = plt.subplots(3, figsize=(10,8), sharex=False, sharey=True)

axs[0].plot(S21wave,S21,drawstyle='steps-mid',color='black',label="Spitzer Spectra")
axs[0].plot(J21wave,J21,drawstyle='steps-mid',linestyle='dashdot',color='red',label="JWST Spectra")
axs[0].set_xlim(11,18)

axs[1].plot(S21wave,S21,drawstyle='steps-mid',color='black',label="Spitzer Spectra")
axs[1].plot(J21wave,J21,drawstyle='steps-mid',linestyle='dashdot',color='red',label="JWST Spectra")
axs[1].set_xlim(18,25)

axs[2].plot(S21wave,S21,drawstyle='steps-mid',color='black',label="Spitzer Spectra")
axs[2].plot(J21wave,J21,drawstyle='steps-mid',linestyle='dashdot',color='red',label="JWST Spectra")
axs[2].set_xlim(25,32)

fig.suptitle("Comparison of Simulated Spitzer and JWST Spectra for Line 21",fontsize = 16)
fig.tight_layout()
fig.subplots_adjust(top=0.93)
legend = axs[0].legend(loc='upper right',fontsize='large')
```

```python
# continuum-subtracted plot
S21_line = Spitzer21['Only_Line[Jy]'].tolist()
J21_line = JWST21['Only_Line[Jy]'].tolist()

#plotting in the 3 panel format
fig, axs = plt.subplots(3, figsize=(10,8), sharex=False, sharey=True)

axs[0].plot(S21wave,S21_line,drawstyle='steps-mid',color='black',label="Spitzer Spectra")
axs[0].plot(J21wave,J21_line,drawstyle='steps-mid',linestyle='dashdot',color='red',label="JWST Spectra")
axs[0].set_xlim(11,18)

axs[1].plot(S21wave,S21_line,drawstyle='steps-mid',color='black',label="Spitzer Spectra")
axs[1].plot(J21wave,J21_line,drawstyle='steps-mid',linestyle='dashdot',color='red',label="JWST Spectra")
axs[1].set_xlim(18,25)

axs[2].plot(S21wave,S21_line,drawstyle='steps-mid',color='black',label="Spitzer Spectra")
axs[2].plot(J21wave,J21_line,drawstyle='steps-mid',linestyle='dashdot',color='red',label="JWST Spectra")
axs[2].set_xlim(25,32)

fig.suptitle("Comparison of Simulated Spitzer and JWST Lines for Line 21",fontsize = 16)
fig.tight_layout()
fig.subplots_adjust(top=0.93)
legend = axs[0].legend(loc='upper right',fontsize='large')
```
