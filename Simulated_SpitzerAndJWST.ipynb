```python
#import the modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Data here

```python
## collecting all the data i'll be using to plot

####JWST

#different temp, same column density
Jdata_T300N18 = pd.read_csv('JWST_Tgas300Ncol1e18.csv')
wave_jN1 =Jdata_T300N18['Lam[um]'].tolist()
line_jN1 = Jdata_T300N18['Line_flux[Jy]'].tolist()

Jdata_T600N18 = pd.read_csv('JWST_Tgas600Ncol1e18.csv')
wave_jN2 = Jdata_T600N18['Lam[um]'].tolist()
line_jN2 = Jdata_T600N18['Line_flux[Jy]'].tolist()

#different column density, same temp
Jdata_T450N17 = pd.read_csv('JWST_Tgas450Ncol1e17.csv')
wave_jT1 = Jdata_T450N17['Lam[um]'].tolist()
line_jT1 = Jdata_T450N17['Line_flux[Jy]'].tolist()

Jdata_T450N19 = pd.read_csv('JWST_Tgas450Ncol1e19.csv')
wave_jT2 = Jdata_T450N19['Lam[um]'].tolist()
line_jT2 = Jdata_T450N19['Line_flux[Jy]'].tolist()
```

```python
####Spitzer

Sdata_T300N18 = pd.read_csv('Spitzer_Tgas300Ncol1e18.csv')
wave_sN1 =Sdata_T300N18['Lam[um]'].tolist()
line_sN1 = Sdata_T300N18['Line_flux[Jy]'].tolist()

Sdata_T600N18 = pd.read_csv('Spitzer_Tgas600Ncol1e18.csv')
wave_sN2 = Sdata_T600N18['Lam[um]'].tolist()
line_sN2 = Sdata_T600N18['Line_flux[Jy]'].tolist()

#different column density, same temp
Sdata_T450N17 = pd.read_csv('Spitzer_Tgas450Ncol1e17.csv')
wave_sT1 = Sdata_T450N17['Lam[um]'].tolist()
line_sT1 = Sdata_T450N17['Line_flux[Jy]'].tolist()

Sdata_T450N19 = pd.read_csv('Spitzer_Tgas450Ncol1e19.csv')
wave_sT2 = Sdata_T450N19['Lam[um]'].tolist()
line_sT2 = Sdata_T450N19['Line_flux[Jy]'].tolist()
```

# Spitzer Plots

## Different temperature, same column density

```python
fig, axs = plt.subplots(3, figsize=(15,8), sharex=False, sharey=False)

axs[0].plot(wave_sN2,line_sN2,drawstyle='steps-mid',color='blue',label="600K")
axs[0].plot(wave_sN1,line_sN1,drawstyle='steps-mid',color='orange',label="300K")
axs[0].set_xlim(11,18)

axs[1].plot(wave_sN2,line_sN2,drawstyle='steps-mid',color='blue')
axs[1].plot(wave_sN1,line_sN1,drawstyle='steps-mid',color='orange')
axs[1].set_xlim(19,25.7)

axs[2].plot(wave_sN2,line_sN2,drawstyle='steps-mid',color='blue')
axs[2].plot(wave_sN1,line_sN1,drawstyle='steps-mid',color='orange')
axs[2].set_xlim(25.8,34)

legend = axs[0].legend(loc='upper left',fontsize='large')
```

## different column density, same temperature

```python
fig, axs = plt.subplots(3, figsize=(15,8), sharex=False, sharey=False)

axs[0].plot(wave_sT2,line_sT2,drawstyle='steps-mid',color='blue',label="Ncol 1e19")
axs[0].plot(wave_sT1,line_sT1,drawstyle='steps-mid',color='orange',label="Ncol 1e17")
axs[0].set_xlim(11,18)

axs[1].plot(wave_sT2,line_sT2,drawstyle='steps-mid',color='blue')
axs[1].plot(wave_sT1,line_sT1,drawstyle='steps-mid',color='orange')
axs[1].set_xlim(19,25.7)

axs[2].plot(wave_sT2,line_sT2,drawstyle='steps-mid',color='blue')
axs[2].plot(wave_sT1,line_sT1,drawstyle='steps-mid',color='orange')
axs[2].set_xlim(25.8,34)

legend = axs[0].legend(loc='upper left',fontsize='large')
```
